```python
def minimum_essai(nb_min:int=1,nb_max:int=10000):
    from numpy import random

    nb_aleatoire = random.randint(nb_min,nb_max)
    essai = 0

    while True:
        nb_choisi = (nb_min+nb_max)//2
        essai += 1
        if nb_choisi == nb_aleatoire:
            return essai
        elif nb_choisi < nb_aleatoire:
            nb_min = nb_choisi
        else:
            nb_max = nb_choisi


print(minimum_essai())
```

```python
def moy_minimum_essai(nb_essai:int = 100000) -> list:
    moyenne = [minimum_essai() for i in range(nb_essai)]
    return moyenne

from numpy import average
print(round(average(moy_minimum_essai()),0))

```

```python
import numpy as np
import matplotlib.pyplot as plt

liste_essai = moy_minimum_essai()

q1 = np.percentile(liste_essai,25)
med = np.median(liste_essai)
q3 = np.percentile(liste_essai,75)
a = np.percentile(liste_essai,20)
b = np.percentile(liste_essai,5)
c = np.percentile(liste_essai,1)
d = np.percentile(liste_essai,10)
e = np.percentile(liste_essai,0.1)

print(f"Q1 = {q1}, mediane = {med}, Q3 = {q3}")
print(f"20% = {a} , 10% = {d} 5% = {b} , 1% = {c} , 0.1% = {e}")


plt.figure(figsize=(10,8))
plt.boxplot(liste_essai,orientation='horizontal')
plt.grid(True)
plt.title("Diagramme moustaches des statistiques d'essais")
plt.xlabel("Nombre")





# si on reussi en moins de 10 essais on gagnes
```

### Stratégie 1
Un joueur doit deviner un nombre entre 1 et 10000 :
- Si le joueur trouve le nombre en **moins de 6 essais ou moins** il mutliplie sa mise par **10**
- Si le joueur trouve le nombre en **8 essais ou moins** il mutliplie sa mise par **8**
- Si le joueur trouve le nombre en **9 essais** il mutliplie sa mise par **5**
- Si le joueur trouve le nombre en **10 essais** il mutliplie sa mise par **4**
- Si le joueur trouve le nombre en **11 essais** il mutliplie sa mise par **2**
- Sinon le joueur perd sa mise

```python
def strategie1(mise:int =10,nb_essai:int =100000): 
    mise_de_depart = mise
    benefices = 0

    list_essais = moy_minimum_essai(nb_essai=nb_essai)
    for essai in list_essais:
        
        if essai <= 6:
            benefices -= mise_de_depart*10 - mise_de_depart
        elif essai <= 8:
            benefices -= mise_de_depart*8 - mise_de_depart
        elif essai <= 9:
            benefices -= mise_de_depart*5 - mise_de_depart
        elif essai <= 10:
            benefices -= mise_de_depart*4 - mise_de_depart
        elif essai <= 11:
            benefices -= mise_de_depart*2 - mise_de_depart
        else:
            benefices += mise_de_depart

    return benefices

print(strategie1())



#moyenne_strategie = average([strategie1(10,10000) for i in range(100)])
#print(f"La moyenne de bénéfice pour 1000 tests de statégie est de {moyenne_strategie} €")
```
